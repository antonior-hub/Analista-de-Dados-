1. Importação das bibliotecas e carregamento do dataset

```python
# Importação das bibliotecas necessárias
import pandas as pd

# Carregamento do dataset Melbourne Housing Snapshot
file_path = '/content/drive/MyDrive/melb_data.csv'
df = pd.read_csv(file_path)

# Visualização das primeiras linhas para verificar os dados
print("Primeiras 5 linhas do dataset:")
df.head()
```

2. Estrutura do Dataset

```python
# Verificação da estrutura do dataset
print("Informações gerais sobre o dataset:")
print(f"Número de linhas: {df.shape[0]}")
print(f"Número de colunas: {df.shape[1]}")
print("\nNomes das colunas:")
print(df.columns.tolist())
```

3. Criação da variável derivada

```python
# Criação da variável derivada: TotalRooms
# Esta variável representa o total de quartos por propriedade,
# somando os quartos principais (Rooms) com os quartos secundários (Bedroom2)

df['TotalRooms'] = df['Rooms'] + df['Bedroom2']

print("Variável TotalRooms criada com sucesso!")
print(f"Estatísticas descritivas da nova variável:")
print(df['TotalRooms'].describe())
```

4. Cálculo do coeficiente de correlação

```python
# Cálculo do coeficiente de correlação de Pearson entre TotalRooms e Price
correlation = df['TotalRooms'].corr(df['Price'])

print(f"Coeficiente de correlação entre TotalRooms e Price: {correlation:.3f}")
```

5. Análise e interpretação dos resultados

```python
# Análise da correlação encontrada
print("ANÁLISE DOS RESULTADOS:")
print("=" * 50)
print(f"Coeficiente de correlação: {correlation:.3f}")
print()

if abs(correlation) >= 0.7:
    intensidade = "forte"
elif abs(correlation) >= 0.3:
    intensidade = "moderada"
else:
    intensidade = "fraca"

if correlation > 0:
    direcao = "positiva"
else:
    direcao = "negativa"

print(f"A correlação é {intensidade} e {direcao}.")
```

6. Conclusão

A variável TotalRooms (soma de Rooms + Bedroom2) apresentou um coeficiente de correlação de 0,493 com a variável alvo Price.
Interpretação:


*   Esse valor indica uma correlação positiva moderada entre o total de
    quartos e o preço dos imóveis;
*   Imóveis com maior número total de quartos tendem a apresentar preços mais elevados, pois mais cômodos agregam valor habitacional.

*   O coeficiente de 0,493 sugere que cerca de 24% da variação nos preços pode ser explicada pelo número total de quartos (R^2 ≈ 0,493^2 ≈ 0,24).


Esta análise confirma que o TotalRooms é um preditor relevante para a estimação do preço dos imóveis no mercado de Melbourne.
